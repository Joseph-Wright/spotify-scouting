```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.oauth2 import SpotifyOAuth
import random
import json
import os
from time import time

# TODO: New artists filter
# TODO: Web app
```

```python
CLIENT_ID = os.env.get("CLIENT_ID")
CLIENT_SECRET = os.env.get("CLIENT_SECRET")
MARKET = "GB"
GENRES = ["indie-pop", "indie", "alt-rock", "alternative"]
SEED_TRACKS = []
SEED_PLAYLIST = "https://open.spotify.com/playlist/03w9EPuK6oCz7VaKd2wM0U?si=374cf2f2a4b94107"
OUTPUT_PLAYLIST = "https://open.spotify.com/playlist/2kWTxfXaMscnx0EFq87phv?si=20ea8a00aa90433b"
TARGET_POPULARITY = 0
```

```python
# sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET))
sp = spotipy.Spotify(
    auth_manager=SpotifyOAuth(
        client_id=CLIENT_ID, 
        client_secret=CLIENT_SECRET, 
        redirect_uri="https://www.google.com",
        scope="playlist-modify-private"
    )
)
```

```python
seed_tracks = sp.playlist_items(SEED_PLAYLIST)

seed_artists = []
for i in seed_tracks["items"]:
    seed_artists.append(i["track"]["artists"][0]["uri"].split(":")[-1])

seed_tracks = [t["track"]["external_urls"]["spotify"] for t in seed_tracks["items"]]
```

```python
available_markets = sp.country_codes
available_genres = sp.recommendation_genre_seeds()["genres"]

assert MARKET in available_markets
for g in GENRES:
    if g not in available_genres:
        print(f"{g} is invalid")
```

```python
playlist = OUTPUT_PLAYLIST
playlist_items = sp.playlist_items(playlist)
existing_artists = set()
for track in playlist_items["items"]:
    existing_artists.update([a["id"] for a in track["track"]["artists"]])
```

```python
for x in range(10):
    results = sp.recommendations(
        seed_tracks=seed_tracks, 
#         seed_genres=GENRES, 
#         seed_artists=seed_artists,
        target_acousticness=random.randint(0,100)/100, 
        target_liveness=0, 
        target_danceability=random.randint(0,100)/100,
        target_valence=random.randint(0,100)/100,
        target_energy=random.randint(0,100)/100,
        target_popularity=0, 
        limit=100, 
        country=MARKET
    )
    tracks = []
    print(len(results["tracks"]), "recommendations")
    for result in results["tracks"]:
        artist_id = result["artists"][0]["id"]
        artist = sp.artist(artist_id)
        name = result["artists"][0]["name"]
        followers = artist["followers"]["total"]
        link = result["external_urls"]["spotify"]
#         print("doing stuff")
        if followers < 1500 and artist_id not in existing_artists:
            print(name)
            print(followers, "followers")
            print(link)
            print()
            tracks.append(link)
            existing_artists.add(artist_id)
    if tracks:
        sp.playlist_add_items(playlist, tracks)
```
